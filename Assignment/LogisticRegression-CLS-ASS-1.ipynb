```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
indep1=dataset.drop(columns=['classification'])
```

```python
dep=dataset['classification']
dep
```

```python
indep = pd.get_dummies(indep1, drop_first=True)
indep
```

```python
dataset['classification'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV
param_grid =  {'solver':['liblinear', 'newton-cg', 'saga', 'lbfgs'],
                          'penalty':['l2'], 'C': [0.1, 1, 10]}
grid = GridSearchCV(LogisticRegression(), param_grid, verbose = 3, refit = True, n_jobs = -1, scoring = 'accuracy')

grid.fit(x_train, y_train)
```

```python
re = grid.cv_results_
grid_predictions = grid.predict(x_test)


from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)

from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)

print(cm)
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, grid_predictions, average = 'weighted')
print("the f1_macro value for best parameter {}: " . format(grid.best_params_), f1_macro)  
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
```
